```python
import numpy as np
from numba import jit
import cProfile, pstats, io
from pstats import SortKey


def printStats( func ):
    def wrapped( *args, **kwargs ):
        pr = cProfile.Profile()
        pr.enable()
        func( *args, **kwargs)
        pr.disable()
        s = io.StringIO()
        sortby = SortKey.CUMULATIVE
        ps = pstats.Stats(pr, stream=s).sort_stats(sortby)
        ps.print_stats()
        print(s.getvalue())
    
    return wrapped
```

```python
class Joint:
    def __init__( length : float, q : float, parent ):
        self.length = length
        self.q = q
        self.parent = parent
        
```

```python
@jit
def fib( n : int ):
    if n == 0:
        return 1
    if n == 1:
        return 1
    else:
        r = fib( n - 1 ) + fib( n - 2 )
        return r
```

```python
pr = cProfile.Profile()
pr.enable()
fib(35)
pr.disable()
s = io.StringIO()
sortby = SortKey.CUMULATIVE
ps = pstats.Stats(pr, stream=s).sort_stats(sortby)
ps.print_stats()
print(s.getvalue())
```
